## Feature exposure

Make binary categories from groups.

```python
df = pd.get_dummies( df, columns = ['Vehicle Size',  'Vehicle Style', 'Driven_Wheels'] )
df.info()
```

```python
df['Market Category'].unique()
```

```python
df = df.drop(['Make', 'Market Category', 'Engine Fuel Type', 'Transmission Type', 'Model', 'highway MPG', 'city mpg', 'Popularity'], axis=1)
df = pd.get_dummies( df, columns = ['Country',  'Market'] )
df = df.dropna()
df.info()
```

```python
df = df[df['MSRP'] > 10000]
df.info()
```

```python


X = df.drop(['MSRP'], axis=1)
X.describe()
```

```python
Y = df['MSRP']
Y.describe()
```

```python
import numpy as np
Xx = np.array(X.values) 
Yy = np.array(Y.values / 100)
```

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam, SGD
from keras.layers.core import Dense, Activation
from keras import optimizers
```

## Create deep learning model

```python
model = Sequential()

model.add(Dense(20, input_dim=np.shape(Xx)[1], activation='sigmoid')) # sigmoid, linear, relu, tanh
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='linear'))

model.compile(Adam(), 'mean_squared_error', metrics=['accuracy'])

model.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xx, Yy, test_size=0.1)
print(np.shape(X_train), np.shape(y_train))

print(X_train[:10])

y_train = np.expand_dims(y_train, axis=1)
print(y_train[:10])
```

```python
model.fit(X_train, y_train, epochs=250)
```

## Test model with untouched data

```python
import math

for i in range(15):
    pred = int(model.predict(np.expand_dims(X_test[i].T, axis=0))[0][0]*100)
    act = int(y_test[i]*100)
    error = float(math.fabs(act - pred) / act) * 100
    print(pred, act, error)
    
```

```python

model_json = model.to_json()
with open("lsci_cars.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("lsci_cars.h5")
print("Saved model to disk")
```
